## Creating function to change the "Value" from string to float

```python
def converter(x, index_of_column_to_convert):
    val=x.iloc[:,index_of_column_to_convert]
    remove_unit = []
    
    for i in range(val.shape[0]):
        remove_unit.append(val[i].strip("€"))
    
    remove_multiplier_K = []
    
    for i in range(len(remove_unit)):
        remove_multiplier_K.append(remove_unit[i].strip("K"))
    
    remove_multiplier_M = []
    
    for i in range(len(remove_multiplier_K)):
        remove_multiplier_M.append(remove_multiplier_K[i].strip("M"))
    
    val_conversion = np.asarray(remove_multiplier_M)
    val_conversion = val_conversion.astype(float)
    
    for i in range(len(val_conversion)):
        if remove_unit[i][-1]=="K":
            val_conversion[i]*=1000
        if remove_unit[i][-1]=="M":
            val_conversion[i]*=1000000
            
    return val_conversion
```

## Applying One Hot Encoding

```python
def encoding(wr):
    low_wr=[]
    med_wr=[]
    high_wr=[]
    
    for i in range(len(awr)):
        if wr[i] == "H":
            low_wr.append(0)
            med_wr.append(0)
            high_wr.append(1)
            
        if wr[i] == "M":
            low_wr.append(0)
            med_wr.append(1)
            high_wr.append(0)
            
        if wr[i] == "L":
            low_wr.append(1)
            med_wr.append(0)
            high_wr.append(0)
            
    low_wr = np.asarray(low_wr)
    med_wr = np.asarray(med_wr)
    high_wr = np.asarray(high_wr)
    
    encod_wr = pd.DataFrame(data=low_wr, columns=["Low"])
    encod_wr["Medium"] = pd.DataFrame(data=med_wr)
    encod_wr["High"] = pd.DataFrame(data=high_wr)
    
    return encod_wr
```

```python
low_awr=[]
med_awr=[]
high_awr=[]

for i in range(len(awr)):
    if awr[i] == "H":
        low_awr.append(0)
        med_awr.append(0)
        high_awr.append(1)
        
    if awr[i] == "M":
        low_awr.append(0)
        med_awr.append(1)
        high_awr.append(0)
        
    if awr[i] == "L":
        low_awr.append(1)
        med_awr.append(0)
        high_awr.append(0)
        
low_awr = np.asarray(low_awr)
med_awr = np.asarray(med_awr)
high_awr = np.asarray(high_awr)
```

```python
encod_awr = pd.DataFrame(data=low_awr, columns=["Low"])
encod_awr["Medium"] = pd.DataFrame(data=med_awr)
encod_awr["High"] = pd.DataFrame(data=high_awr)
```
